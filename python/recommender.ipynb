```python
df['squareFeet'] = df['squareFeet'].str.replace("sqft", "", regex=True).astype(float)
x = df.drop(['_id', 'description', 'budget', 'state', 'city', 'floor', 'furnishing', 'propertyType', 'facing', 'address'], axis=1)
dms = pd.get_dummies(df[['state', 'city', 'floor', 'furnishing', 'propertyType', 'facing']], drop_first=True)
X = pd.concat([x, dms], axis=1)
Y = df['budget']
X.fillna(X.mean(), inplace=True)  # Replace NaN with column mean
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.35,random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(x_train)
X_test = scaler.transform(x_test)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
from xgboost import XGBRegressor
xgb = XGBRegressor().fit(X_train,y_train)
y_pred = xgb.predict(X_test)
print(np.sqrt(mean_squared_error(y_pred,y_test)))
```

```python
print(y_pred[0])
```

```python
from sklearn.svm import SVR
svr = SVR(kernel='poly').fit(X_train,y_train)
svr_pred = svr.predict(X_test)
print(np.sqrt(mean_squared_error(y_test,svr_pred)))
```

```python
print(svr_pred[0])
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor(n_neighbors=3).fit(X_train,y_train)
knn_pred = knn.predict(X_test)
print(np.sqrt(mean_squared_error(y_test,knn_pred)))
```
